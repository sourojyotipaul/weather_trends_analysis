```python
#finally checking the state of the dataframe
df.info()
```

## Exploratory Data Analysis
Now we have clean dataframe on which we can now perform certain operations to get the answer of our desired querries.

```python
#introducing new column release_year by formatting existing column release_date
df["release_year"] = df["release_date"].dt.strftime('%Y')
```

```python
df.release_year.value_counts().sort_values().plot(kind = 'line', figsize=(20,10), title='Number of movies per year VS Year')
plt.xlabel('Year')
plt.ylabel('Number Of Movies Per Year')
```

From the above graph we can say that from 1960 onwards the number of movies per year has increased significantly.
Hence it will be appropriate to take 1960 as a base year, therefore removing all the other rows before release year 1960.

```python
#taking backup of the dataframe
df_backup = df.copy()
```

```python
#converting the datatype of release_year from string to int
df['release_year'] = df['release_year'].astype(int)

#modifying the dataframe, keeping records of movies from release_year 1960
df = df[df.release_year > 1959]
```

```python
#defining function to calculate mean. Code reusability
def group_by_avg(feature1,feature2):
    gba = df.groupby(feature1)[feature2].mean()
    return gba
```

```python
#let us find for the first question which we have.
#How movies are doing over time, How are old movies rated against the new ones.

avg_rating = group_by_avg("release_date","vote_average")
```

```python
avg_rating.head()
```

```python
avg_rating.plot(kind='line', title='Average rating VS Release Date', figsize=(20,10))
```

```python
#From the above line plot the it is very difficult to infer anything as the graph is very dense towards the right.
#Hence we need to plot graph with release year 
```

```python
avg_rating_per_year = group_by_avg("release_year","vote_average")
```

```python
avg_rating_per_year.plot(kind='line', title='Average rating per year VS Release Date', figsize=(20,10))
#x-axis name
plt.xlabel('Year')
#y-axis name
plt.ylabel('Average Rating')
```

### Observation 1
##### (Above Graph : Figure1)
As far as rating has concerned, that has gradually decreased over the year as we can see from the graph.
Which means old movies were rated much higher than the new ones

#### Clearly number of movies per year have increased significantly over the years

```python
#let us now see how is the film industry performing in terms of profit or loss over the years
#Adding a now column 'Profit' to check if there has been a profit or a loss by performing subtract operation 
#Profit = Revenue - Budget
#Now if the result is negative that means there has been a loss where as for positive results it means profit
```

```python
df.insert(1,"profit", df["revenue"] - df["budget"])
df.head()
```

```python
#here we will group by release_year and calculate the mean of profit for each year
```

```python
avg_profit_per_year = group_by_avg('release_year','profit')
```

```python
avg_profit_per_year.plot(kind='line', color='green', title='Profit Average VS Year', figsize=(20,10))
#x-axis name
plt.xlabel('Year')
#y-axis name
plt.ylabel('Profit Average')
```

```python
#scatter plot to understand the relation
plt.figure(figsize=(20,20), dpi=100)
sns.regplot(x=avg_profit_per_year.index, y=avg_profit_per_year).set_title('Profit VS Release Year', size=25)
```
